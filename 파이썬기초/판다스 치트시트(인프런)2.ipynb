```python
df = sns.load_dataset('iris')
df.head()
```

```python
columns = ['sepal_width','sepal_length','species']
df[columns]
```

```python
#regex="  "  지정한문자가 들어간칼럼 출력
df.filter(regex="_")

# df.filter(regex='length$') #length로 끝나는 칼럼 출력
# df.filter(regex='^Sepal') # Sepal로 시작하는 칼럼 출력
```

```python
df.filter(regex='^x[1-5]$') # 숫자,영문자 섞여있는경우
```

```python
df.filter(regex='^(?!species$).*') # species라는 문자가들어간칼럼 제외하고 출력
```
