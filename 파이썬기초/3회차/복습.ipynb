### 데이터 타입 변경
#### 분양가격이 object타입으로 되어있음. 수치형데이터로 변경. 결측치가있으면 변환제대로 안됨 
#### pd.to_numeric으로 타입변경

```python
type(np.nan)
```

```python
# errors='coerce' 유효하지않은값을 강제로 NaN값으로 처리

df_last['분양가격'] = pd.to_numeric(df_last['분양가격(㎡)'], errors='coerce')
df_last['분양가격'].head()
```

### 평당분양가격 구하기

```python
df_last['평당분양가격'] = df_last['분양가격'] * 3.3
df_last.head()
```

### 규모구분을 전용면적 칼럼으로 변경

```python
df_last['규모구분'].unique()
```

```python
df_last['전용면적'] = df_last['규모구분'].str.replace("전용면적","")
df_last['전용면적'] = df_last['전용면적'].str.replace("초과","~")
df_last["전용면적"] = df_last["전용면적"].str.replace("이하","")
df_last["전용면적"] = df_last["전용면적"].str.replace(" ","").str.strip()
df_last["전용면적"].unique()
```

### groupby로 데이터집계하기

```python
# 지역명으로 분양가격의 평균을 구하고 막대그래프(bar)로 시각화
# df.groupby(["인덱스로 사용할 칼럼명"])["계산할 칼럼값"].연산()

df_last.groupby(["지역명"])["평당분양가격"].mean()
```

```python
#연도, 지역명으로 평당분양가격의 평균을 구한다

g = df_last.groupby(["연도","지역명"])["평당분양가격"].mean()
g
```

### 최근 데이터 시각화 하기

### lmplot & swarmplot

```python
df_last["연도"].unique()
```

```python
sns.regplot(data=df_last,x="연도",y="평당분양가격")
```

지터Jitter는 데이터 값에 약간의 노이즈를 추가하는 방법을 말한다. 
노이즈를 추가하면 데이터 값이 조금씩 움직여서 같은 값을 가지는 
데이터가 그래프에 여러 번 겹쳐서 표시되는 현상을 막아준다

```python
sns.regplot(data=df_last,x="연도",y="평당분양가격",x_jitter=.1)
```

```python
#연도별 평당분양가격을 lmplot으로 그리기
# hue옵션으로 전용면적을 표시

sns.lmplot(data=df_last, x="연도",y="평당분양가격",hue="전용면적",col="전용면적",
           col_wrap=3, x_jitter=.1)
```

```python
# 연도별 평당분양가격을 swarmplot으로 그리기
# swarmplot은 범주형 데이터의 산점도를 표현하기에 적합

plt.figure(figsize=(10,4))
sns.swarmplot(data=df_last,x="연도",y="평당분양가격",hue="전용면적",size=1)
```
