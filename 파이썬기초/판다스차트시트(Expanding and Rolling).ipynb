```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000))
s.plot()
```

```python
s = s.cumsum()
s.plot()
```

```python
r = s.rolling(window=60) #
r.mean()
```

```python
#   

s.plot(style='k--')
r.mean().plot(style='k--')
```
